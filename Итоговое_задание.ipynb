```python
data_onehot = pd.DataFrame({'robot': [1 if x == 'robot' else 0 for x in data['whoAmI']],
                            'human': [1 if x == 'human' else 0 for x in data['whoAmI']]})
data_onehot.head()
```

```python
lst = ['robot'] * 10 + ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})
data = pd.get_dummies(data, columns=['whoAmI'])
data.head()
```
